````python
# ==========================================
# LUMINA AI SERVER - THREADED FIX
# ==========================================

# 1. INSTALL DEPENDENCIES
!pip install -q fastapi uvicorn pyngrok nest-asyncio torch transformers accelerate

import torch
from transformers import AutoModelForCausalLM, AutoTokenizer, pipeline
from fastapi import FastAPI
from pydantic import BaseModel
from pyngrok import ngrok
import uvicorn
import nest_asyncio
import json
import threading

# 2. CONFIGURATION
# PASTE YOUR NGROK TOKEN BELOW BEFORE RUNNING
NGROK_TOKEN = "37IDM9muiUKQsBCLFtnU2H7kzDF_aGuCeKCXitZUwLzCkNuW"
MODEL_ID = "Qwen/Qwen2.5-1.5B-Instruct"

# 3. LOAD AI MODEL
print("⏳ Loading AI Model...")
tokenizer = AutoTokenizer.from_pretrained(MODEL_ID)
model = AutoModelForCausalLM.from_pretrained(
    MODEL_ID,
    torch_dtype=torch.float16,
    device_map="auto"
)

pipe = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    max_new_tokens=512,
    temperature=0.7,
    top_p=0.9,
    do_sample=True,
)
print("✅ AI Model Loaded!")

# 4. DEFINE API
app = FastAPI()

class PlaylistRequest(BaseModel):
    userId: str
    context: str
    history: list[str] = []

@app.get("/")
def home():
    return {"status": "Lumina AI Online", "model": MODEL_ID}

@app.post("/generate")
def generate_playlist(req: PlaylistRequest):
    print(f"📩 Received request: {req.context}")
    history_text = ", ".join(req.history) if req.history else "Pop, Viral Hits"

    messages = [
        {"role": "system", "content": "You are a music DJ API. You output ONLY valid raw JSON arrays of strings (Artist - Song). No markdown."},
        {"role": "user", "content": f"User Taste: {history_text}. Create a playlist of 8 songs for vibe: '{req.context}'. Output JSON:"}
    ]

    prompt = tokenizer.apply_chat_template(messages, tokenize=False, add_generation_prompt=True)
    outputs = pipe(prompt)
    response_text = outputs[0]["generated_text"].split("<|im_start|>assistant")[-1].strip()
    response_text = response_text.replace("```json", "").replace("```", "").strip()

    try:
        return json.loads(response_text)
    except:
        return []

# 5. START SERVER (THREADED TO FIX RUNTIME ERROR)
ngrok.set_auth_token(NGROK_TOKEN)
public_url = ngrok.connect(8000).public_url
print(f"\n🚀 LUMINA AI SERVER RUNNING AT: {public_url}\n")
print(f"👉 Copy this URL into services/aiService.ts\n")

nest_asyncio.apply()

# Run uvicorn in a separate thread so it doesn't block Colab's loop
def run_server():
    uvicorn.run(app, port=8000)

thread = threading.Thread(target=run_server)
thread.start()
````
